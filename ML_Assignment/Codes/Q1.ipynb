3.Ploting histograms for continuous valued attributes and bar graphs for the discrete

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data['sepal length'], color='green')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Count")
plt.title("histogram of Sepal Length" )
```

```python
plt.hist(data['sepal width'], color='blue')
plt.xlabel("Sepal width (cm)")
plt.ylabel("Count")
plt.title("histogram of Sepal width" )
```

```python
#ploting bar graph betwwn sepal lenth and target class
plt.bar(data['class'],data['sepal length'],color = ['teal','aqua','steelblue']) 
plt.xlabel("class")
plt.ylabel("sepal length(cm)")
plt.title('Bar plot of iris class/species vs sepal length')
plt.show()
```

```python
#ploting bar graph betwwn sepal width and target class
plt.bar(data['class'],data['sepal width'],color = ['brown','firebrick','maroon']) 
plt.xlabel("class")
plt.ylabel("sepal width(cm)")
plt.title('Bar plot of iris class/species vs sepal width')
plt.show()
```

```python
#ploting bar graph betwwn Petal lenth and target class
plt.bar(data['class'],data['petal length'],color = ['lime','seagreen','springgreen']) 
plt.xlabel("class")
plt.ylabel("petal lenth(cm)")
plt.title('Bar plot of iris class/species vs petal length')
plt.show()
```

```python
#ploting bar graph betwwn Petal width and target class
plt.bar(data['class'],data['petal width'],color = ['bisque','orange','tan']) 
plt.xlabel("class")
plt.ylabel("petal width(cm)")
plt.title('Bar plot of iris class/species vs petal width')
plt.show()
```

Using TSNE (t-distributed stochastic neighbour embedding) algorithm to reduce
data dimensions to 2

```python
from sklearn.manifold import TSNE
```

```python
model = TSNE(n_components=2, random_state=0)
```

```python
MNIST_Data = MNIST_Data.reshape(MNIST_Data.shape[0], -1)
tsne_data = model.fit_transform(MNIST_Data)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(tsne_data[:,0], tsne_data[:,1], c=MNIST_Data_Labels, alpha=0.5)
plt.xlabel('t-SNE-1')
plt.ylabel('t-SNE-2')
```
